### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20

# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint

# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# Assign the 'uk_food' database to a variable name
db = mongo['uk_food']

# Assign the 'establishments' collection to a variable
establishments = db['establishments']

# Define the query to find establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
first_document = establishments.find_one(query)
if first_document:
    print("First document with a hygiene score of 20:")
    pprint(first_document)
else:
    print("No document found with a hygiene score of 20.")


```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame

# Import dependencies
from pymongo import MongoClient
import pandas as pd

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# Assign the 'uk_food' database to a variable name
db = mongo['uk_food']

# Assign the 'establishments' collection to a variable
establishments = db['establishments']

# Define the query to find establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Find documents that match the query
cursor = establishments.find(query)

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(cursor))

# Display the number of rows in the DataFrame
row_count = len(df)
print(f"Number of rows in the DataFrame: {row_count}")

# Display the first 10 rows of the DataFrame
first_10_rows = df.head(10)
print("First 10 rows of the DataFrame:")
print(first_10_rows)

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# Assign the 'uk_food' database to a variable name
db = mongo['uk_food']

# Assign the 'establishments' collection to a variable
establishments = db['establishments']

# Define the latitude and longitude for "Penang Flavours"
latitude = 51.4820  # Replace with the actual latitude
longitude = 0.0077  # Replace with the actual longitude

# Define the search radius in degrees
degree_search = 0.01

# Define the query to find establishments with RatingValue of 5 and within the search radius
query = {
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
    'RatingValue': 5
}

# Define the sorting criteria (ascending hygiene score)
sort = [('scores.Hygiene', 1)]

# Find the top 5 establishments that match the query and are sorted by hygiene score
results = establishments.find(query).sort(sort).limit(5)

# Print the results using both print and pprint
print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score and nearest to 'Penang Flavours':")
for idx, establishment in enumerate(results, start=1):
    print(f"{idx}. {establishment['BusinessName']} (Hygiene Score: {establishment['scores']['Hygiene']})")
    pprint(f"{idx}. {establishment['BusinessName']} (Hygiene Score: {establishment['scores']['Hygiene']})")
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Import necessary libraries
from pymongo import MongoClient

# Create a MongoDB client
client = MongoClient("mongodb://localhost:27017/")

# Connect to your database
db = client['uk_food']

# Define the aggregation pipeline
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0  # Match establishments with hygiene score of 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",  # Group by Local Authority Name
            "count": {"$sum": 1}  # Count the matching documents in each group
        }
    },
    {
        "$sort": {
            "count": -1  # Sort in descending order of count
        }
    }
]

# Execute the aggregation pipeline
results = list(db.establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of establishments with a hygiene score of 0 in each Local Authority:")
for result in results:
    print(f"{result['_id']}: {result['count']} establishments")

# Print the first 10 results
print("\nFirst 10 results:")
for result in results[:10]:
    print(f"{result['_id']}: {result['count']} establishments")

```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame

# Import necessary libraries
from pymongo import MongoClient
import pandas as pd

# Create a MongoDB client
client = MongoClient("mongodb://localhost:27017/")

# Connect to your database
db = client['uk_food']

# Define the aggregation pipeline
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0  # Match establishments with hygiene score of 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",  # Group by Local Authority Name
            "count": {"$sum": 1}  # Count the matching documents in each group
        }
    },
    {
        "$sort": {
            "count": -1  # Sort in descending order of count
        }
    }
]

# Execute the aggregation pipeline
results = list(db.establishments.aggregate(pipeline))

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
row_count = len(df)
print(f"Number of rows in the DataFrame: {row_count}")

# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(df.head(10))

```
