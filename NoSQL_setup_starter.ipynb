5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Define the query to select all documents in the collection
all_documents_query = {}

# Retrieve all documents in the collection
all_documents = establishments.find(all_documents_query)

# Iterate through each document and update the data type of latitude and longitude
for document in all_documents:
    # Get the current latitude and longitude as strings
    latitude_str = document['geocode']['latitude']
    longitude_str = document['geocode']['longitude']
    
    # Convert the latitude and longitude strings to float
    latitude_float = float(latitude_str)
    longitude_float = float(longitude_str)
    
    # Update the document with the new data types
    update_query = {'_id': document['_id']}
    update_data = {
        '$set': {
            'geocode.latitude': latitude_float,
            'geocode.longitude': longitude_float
        }
    }
    
    # Update the document in the collection
    update_result = establishments.update_one(update_query, update_data)
    
    if update_result.modified_count > 0:
        print(f"Document {document['_id']} updated with Decimal latitude and longitude.")
    
print("Data type conversion complete.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, {'$set': {"RatingValue": None}})
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many(
    {"RatingValue": {"$exists": True, "$type": "string"}},
    [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}]
)
```

```python
# Check that the coordinates and rating value are now numbers
# Define a query to find documents where latitude, longitude, and RatingValue are numbers
numeric_query = {
    "geocode.latitude": {"$type": "number"},
    "geocode.longitude": {"$type": "number"},
    "RatingValue": {"$type": "number"},
}

# Count the number of documents that match the query
numeric_count = establishments.count_documents(numeric_query)

print(f"Number of documents with numeric latitude, longitude, and RatingValue: {numeric_count}")
```
